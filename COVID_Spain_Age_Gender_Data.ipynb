## Import Packages

```python
import pandas as pd
import numpy as np
```

## Read HTML from link

```python
data_spain_wikipedia = pd.read_html('https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Spain')
```

```python
women_Data_spain = data_spain_wikipedia[9]
```

```python
men_Data_spain = data_spain_wikipedia[10]
```

```python
women_Data_spain['Gender'] = 'Female'
```

```python
men_Data_spain['Gender'] = 'Male'
```

```python
women_Data_spain
```

```python
gender_spain_data = women_Data_spain.append(men_Data_spain)
```

```python
gender_spain_data
```

```python
gender_spain_data.reset_index(inplace=True)
```

```python
gender_spain_data = gender_spain_data.drop(['index'], axis = 1)
```
